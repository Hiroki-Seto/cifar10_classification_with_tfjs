```python
import numpy as np
import tensorflowjs as tfjs
```

```python
from keras.models import load_model
from keras.utils.vis_utils import plot_model
from keras.datasets import cifar10
```

```python
DATA_DIR = "./data"
```

```python
if False:
    (X_train, y_train), (X_test, y_test) = cifar10.load_data()

    if not os.path.exists(DATA_DIR):
        os.makedirs(DATA_DIR)
    np.save(os.path.join(DATA_DIR, "X_train.npy"), X_train)
    np.save(os.path.join(DATA_DIR, "y_train.npy"), y_train)
    np.save(os.path.join(DATA_DIR, "X_test.npy"), X_test)
    np.save(os.path.join(DATA_DIR, "y_test.npy"), y_test)
else:
    X_train = np.load(os.path.join(DATA_DIR, "X_train.npy"))
    y_train = np.load(os.path.join(DATA_DIR, "y_train.npy"))
    X_test = np.load(os.path.join(DATA_DIR, "X_test.npy"))
    y_test = np.load(os.path.join(DATA_DIR, "y_test.npy"))
```

```python
# preprocess
mean = X_train.mean(axis=(0, 1, 2))
std = X_train.std(axis=(0, 1, 2))
X_test = (X_test - mean) / std
```

```python
# keras load model
keras_model = load_model("keras_model/type4.hdf5")
```

```python
# model prediction
pred_prob = keras_model.predict(X_test[:100])
```

```python
# accuracy calculation
pred = np.argmax(pred_prob, axis=1)
true = y_test[:100].ravel()

acc = np.sum(true == pred) / len(pred)

print(acc)
```

```python
SAVE_DIR = "../tfjs_model"
```

```python
if not os.path.exists(SAVE_DIR):
    os.path.exists(SAVE_DIR)
```

```python
tfjs.converters.save_keras_model(keras_model, os.path.join(SAVE_DIR, "wide-resnet.json"))
```
